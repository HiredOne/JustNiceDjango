```python
unique_ingr = {}
```

```python
# for x in list(df["Ingredients"][:10]):
#     print(type(x))
```

```python
index = 1

for row in list(df["Ingredients"][:]):
    print(index, type(row))
    ingredients = row.split(",")
    for ingredient in ingredients:
        if ingredient not in unique_ingr:
            unique_ingr[ingredient] = 0
        unique_ingr[ingredient] = unique_ingr[ingredient] + 1
    index+=1
```

```python
len(unique_ingr)
```

```python
final = pd.DataFrame.from_dict(unique_ingr, orient="index", columns=["Count"])
```

```python
final = final.reset_index().rename(columns={"index": "Ingredient"})
```

```python
final["Ingredient"] = final["Ingredient"].map(lambda x: str.lower(x))
```

```python
final["sort"] = final["Ingredient"].str.extract('(\d+)',expand=False)
```

```python
final.fillna(0, inplace=True)
```

```python
final["sort"] = final["sort"].astype(int)
```

```python
final = final.sort_values(["sort", "Ingredient"] , ascending=[True, True]).reset_index(drop=True)
```

```python
final.drop("sort", axis=1, inplace=True)
```
